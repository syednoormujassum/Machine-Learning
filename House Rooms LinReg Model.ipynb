```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,7
```

```python
X=np.random.rand(100,1)+4
X
```

```python
X.shape
```

```python
y = np.random.randn(100,1)+8*X+579
y
```

```python
plt.plot(X,y,'r^', color='blue')
plt.xlabel('Rooms')
plt.ylabel('Price in Thousands CAD $')
plt.show()
```

From the Plot we can see that as the number of rooms increases the price also increases linearly

```python
LinReg = LinearRegression()
LinReg.fit(X,y)
print('Linear Regression intercept = ', LinReg.intercept_, ' and Linear Regression Co-efficient = ', LinReg.coef_)
```

```python
print('Score = ',LinReg.score(X,y))
```
