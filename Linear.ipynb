```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize']= 10,8
```

```python
rooms = 2*np.random.rand(100,1)+3
rooms
```

```python
price = 265+6*rooms+abs(np.random.randn(100,1))
price[1:10]
```

```python
plt.plot(rooms,price, 'r^')
plt.xlabel(' Number of Rooms in 2020')
plt.ylabel('2020 Average room price in USD')
plt.show
```

```python
X = rooms
y = price
LinReg = LinearRegression()
LinReg.fit(X,y)
print('Intercept =',LinReg.intercept_, '& Coefficient = ',LinReg.coef_ )
```

```python
print('Score of Regression model is ', LinReg.score(X,y))
```
